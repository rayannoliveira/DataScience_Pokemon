```python
import pandas as pd 
```

```python
df= pd.read_csv('pokemon_data.csv')
df
```

```python
#Find all types of Type 1 and putting to Series Type_of_pokmon
Type_of_pokmon = df['Type 1']
```

```python
#Drop the duplicates
Type_of_pokmon.drop_duplicates(inplace=True)
Type_of_pokmon
```

```python
# the shape of Type_of_pokmon
Type_of_pokmon.shape[0]
```

```python
#Redefind the index 
Type_of_pokmon.index = range(Type_of_pokmon.shape[0])
Type_of_pokmon
```

```python
#How many pokemon has type 1

df['Type 1'].value_counts()
```

```python
df['Type 1'].value_counts(normalize = True) * 100
```

```python
#Frequence and the percentage

frequence = df['Type 1'].value_counts()

percentage = df['Type 1'].value_counts(normalize = True) * 100
```

```python
#creating a DataFrame of frequence and percentage 

df_frequence_qualitative = pd.DataFrame({'Frequence': frequence, 'Percentage (%)': percentage})
df_frequence_qualitative
```

```python
#Rename the column

df_frequence_qualitative.rename_axis('Pokemon Type', axis= 'columns', inplace=True)
df_frequence_qualitative
```

```python
df_frequence_qualitative['Frequence'].plot.bar(width=1, color = 'blue', alpha = 0.2, figsize= (12, 6))
```
